```python
from rules import Hand
from card import Card, Suit
from probability_utils import build_hands, build_dealer_hands

player_hand= Hand([Card(Suit.spades, 10), Card(Suit.spades, 2)])
possible_hands = build_dealer_hands([player_hand])
possible_hands
```

```python
import pandas as pd
import numpy as np
from probability_utils import calculate_prob, Card, ProbabilityDeck
df = pd.DataFrame({'hand': possible_hands})
df['value'] = [hand.value for hand in df.hand]
df['winloss'] = np.nan
df['winloss'][df.value > 21] = -1
deck = ProbabilityDeck(n=6)
df['prob'] = [calculate_prob(hand, [Card.from_desc('2 spades')], deck) for hand in df.hand]
```

```python
df.iloc[0].hand.cards
```

```python
dealer_hands=build_dealer_hands([Hand([Card.from_desc('2 spades')])])
dealer_hand_probs = [calculate_prob(hand, df.iloc[0].hand.cards + [Card.from_desc('2 spades')], deck) for hand in dealer_hands]
dealer_hand_probs
```
